```python
import bs4
```

```python
import requests
```

```python
url="https://www.daraz.pk/products/pubg-mobile-221-uc-unknown-cash-i114366914-s1266142453.html?spm=a2a0e.home.flashSale.2.35e34937jUxZfs&search=1&mp=1&c=fs"
```

```python
data=requests.get(url)
```

```python
soup=bs4.BeautifulSoup(data.text,"html.parser")
```

```python
print(soup.prettify())
```

```python
for para in soup.find("p"):
    print(para)
```

```python
for para in soup.find_all("p"):
    print(para)
```

```python
for para in soup.find_all("p"):
    print(para.text)
```

```python
for links in soup.find_all("a"):
    link =links.get("href")
    print (link)
```

```python
for links in soup.find_all("a"):
    link =str(links.get("href"))
    if link [0:2] == "//"  :
        c=("http:"+ link)
        print (c)    
    elif link [0:2] == "None" :
        print("http:"+ link)
```

```python
for ps in soup.find_all("li"):
    print(ps.text)
```
